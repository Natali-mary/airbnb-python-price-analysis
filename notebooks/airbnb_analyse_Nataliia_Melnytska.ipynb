```python
### Version2: 6.1. Superhost nach Gesamtzahl der Anzeigen nach Städten

# --- 2) Spaltenreihenfolge sicherstellen: False (Host), True (Superhost) ---
if set(pivot_tabele.columns.tolist()) == {False, True}:
    pivot_tabele = pivot_tabele.reindex(columns=[False, True])

# Deutsche Legenden-Beschriftung
legend_labels = ['Host', 'Superhost']  # entspricht [False, True]

# --- 3) Farben entsprechend der Spalten ---
colors = ['#FF5A5F', '#00A699']  # Host (False), Superhost (True)

# --- 4) Gestapeltes Balkendiagramm aus absoluten Werten zeichnen ---
ax = pivot_tabele.plot.bar(stacked=True, figsize=(12, 6), color=colors)

# --- 5) Gestaltung (alles auf Deutsch) ---
ax.set_title('Superhost nach Gesamtzahl der Anzeigen nach Städten', fontsize=16, fontweight='bold')
ax.set_xlabel('Stadt')
ax.set_ylabel('Anzahl der Anzeigen')
plt.xticks(rotation=0)
ax.legend(title='Host-Typ', labels=legend_labels)
plt.grid(True, axis='y', alpha=0.3)
plt.tight_layout()

# --- 6) Prozentanteile JE STADT (Zeilenweise: jede Stadt = 100 %) ---
zeilen_summen = pivot_tabele.sum(axis=1)                               # Summe pro Stadt
prozent_tabele = (pivot_tabele.div(zeilen_summen, axis=0) * 100)       # Anteile innerhalb der Stadt
prozent_tabele = prozent_tabele.replace([np.inf, np.nan], 0)

# --- 7) Prozentwerte im jeweiligen Segment zentriert anzeigen ---
containers = ax.containers  # Reihenfolge entspricht pivot_tabele.columns

for j, container in enumerate(containers):      # j = 0 -> Host(False), j = 1 -> Superhost(True)
    for i, bar in enumerate(container):         # i = Index der Stadt (Zeile)
        hoehe = bar.get_height()
        if hoehe <= 0:
            continue
        x = bar.get_x() + bar.get_width() / 2
        y = bar.get_y() + hoehe / 2
        pct = prozent_tabele.iloc[i, j]         # Anteil für (Stadt i, Kategorie j)
        ax.text(x, y, f'{pct:.0f}%', ha='center', va='center',
                color='white', fontsize=10, fontweight='bold')

plt.show()
```

###
Alle Städte weisen Abweichungen auf, insbesondere jedoch:

- London, Paris, Amsterdam → sehr hohe Preise (bis zu 18.000).


- Die meisten Preise (Boxen) liegen im Bereich von etwa 0 bis 500.
- Der Medianwert liegt in den meisten Städten nahe der Untergrenze (voraussichtlich 50 bis 150).
- Die Ausreißer verzerren den Durchschnittswert stark, daher ist es für die Analyse besser, den Median oder den getrimmten Durchschnitt zu verwenden.

Möglicherweise lohnt es sich, eine Segmentierung vorzunehmen:
- 'Normale Angebote' (bis zu 500).
- 'Luxusangebote' (über 500).

```python
### 8.1 BOXPLOT – Preise in Städten - zwei boxplot nebeneinander, mit und ohne Ausreißern
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[column] >= lower) & (df[column] <= upper)]

mean_order = (df.groupby("city")["total_room_price"].median().sort_values(ascending=False).index)

df_no_outliers = remove_outliers_iqr(df, "total_room_price")

# (optional) Berechnen, wie viele Zeilen gelöscht wurden.
removed_pct = 100 * (1 - len(df_no_outliers) / len(df))

# Visualisierung: zwei sns.boxplot nebeneinander
sns.set(style="whitegrid")

fig, axes = plt.subplots(1, 2, figsize=(20, 8))

# Mit Ausreißern
sns.boxplot(data=df, x="city", y="total_room_price", ax=axes[0], order=mean_order, palette=mein_palette)
axes[0].set_title("Mit Ausreißern")
axes[0].set_xlabel("Stadt")
axes[0].set_ylabel("Preis")
axes[0].tick_params(axis='x', rotation=45)

# Ohne Ausreißern
sns.boxplot(data=df_no_outliers, x="city", y="total_room_price", ax=axes[1], order=mean_order, palette=mein_palette)
axes[1].set_title(f"Ohne Ausreißer (entfernt: {removed_pct:.1f}%)")
axes[1].set_xlabel("Stadt")
axes[1].set_ylabel("Preis")
axes[1].tick_params(axis='x', rotation=45)

plt.suptitle("Preisverteilung nach Stadt (Vergleich)", y=0.98)
plt.tight_layout()
plt.show()
```

```python
## 13.1 Verteilung der Zimmerarten nach Room Type
sns.set_theme(style="whitegrid")

min_idx = df_room_type['count'].idxmin()
explode = [0.0] * len(df_room_type)
explode[min_idx] = 0.08 

colors = ('#ff5a5f', '#00a699', '#812581')

plt.figure(figsize=(6, 6))
wedges, texts, autotexts = plt.pie(
    df_room_type['count'],
    labels=df_room_type['room_type'],
    autopct='%1.1f%%',
    colors=colors,
    startangle=140,
    textprops={'fontsize': 12},
    explode=explode,               
    pctdistance=0.8,            
    labeldistance=1.05          
)

for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold')

plt.gca().set_aspect('equal')

plt.title('Verteilung der Zimmerarten', fontsize=14)

total = df_room_type['count'].sum()
legend_labels = [
    f"{row.room_type}: {row.count:,} ({row.count / total * 100:.1f}%)"
    for row in df_room_type.itertuples(index=False)
]

plt.legend(
    wedges,
    [f"{row.room_type}: {row.count:,} ({row.count / df_room_type['count'].sum() * 100:.1f}%)"
    for row in df_room_type.itertuples(index=False)],
    title="Zimmerarten",
    loc="lower center",
    bbox_to_anchor=(0.5, -0.3),
    ncol=1,
    frameon=True
)

plt.tight_layout()
plt.show()
```

```python
### 17. Korrelationsmatrix für numerische Variablen
plt.figure(figsize=(12, 10))

# Relevante numerische Spalten auswählen
numeric_cols = ['total_room_price', 'person_capacity', 'cleanliness_rating', 'guest_satisfaction_rating',
                '#_bedrooms', 'centr_dist', 'metro_dist', 'longitude', 'latitude']

correlation_matrix = df[numeric_cols].corr()

# Heatmap erstellen
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0,
            square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Korrelations-Heatmap der wichtigsten Variablen', fontsize=14, fontweight='bold', pad=20)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

✅ Wichtige Erkenntnisse:

1. Starke positive Korrelation zwischen den Bewertungen
 - cleanliness_rating und guest_satisfaction_rating korrelieren stark (≈ 0,71).
→ Je sauberer die Unterkunft, desto zufriedener sind die Gäste.

2. Anzahl der Schlafzimmer hängt mit der Kapazität zusammen
 - #_bedrooms und person_capacity haben eine hohe Korrelation (≈ 0,56).
→ Mehr Schlafzimmer bedeuten mehr Platz für Gäste.

3. Zimmerpreis und Preis hat kaum Einfluss auf Bewertungen
 - total_room_price und price_per_person zeigen eine schwache oder negative Korrelation mit Sauberkeit und Zufriedenheit.
→ Hoher Preis garantiert keine bessere Bewertung.

4. Entfernung zum Zentrum und zur Metro hängen zusammen
 - centr_dist und metro_dist korrelieren (≈ 0,56).
→ Je weiter vom Zentrum entfernt, desto weiter von der Metro.

5. Geografische Koordinaten (Longitude, Latitude)
 - Nur geringe Korrelation mit Preis und Bewertungen, aber leichte Abhängigkeit zur Entfernung vom Zentrum (≈ 0,41).
→ Standort beeinflusst die Erreichbarkeit, aber nicht direkt die Qualität.

🔍 Geschäftliche Implikationen:

 - Verbesserung der Sauberkeit → höhere Gästezufriedenheit.
 - Preis ist kein Hauptfaktor für gute Bewertungen – Qualität zählt mehr.
 - Standort spielt eine Rolle für die Zugänglichkeit, aber weniger für die Bewertung.

```python
### 19. Streudiagramm auf Kartenbasis (px.scatter_mapbox)

# 1) Koordinaten der Städte (lat/lon)
city_coords = {
    'London':    {'lat': 51.5074, 'lon': -0.1278},
    'Rome':      {'lat': 41.9028, 'lon': 12.4964},
    'Paris':     {'lat': 48.8566, 'lon': 2.3522},
    'Lisbon':    {'lat': 38.7223, 'lon': -9.1393},
    'Athens':    {'lat': 37.9838, 'lon': 23.7275},
    'Budapest':  {'lat': 47.4979, 'lon': 19.0402},
    'Vienna':    {'lat': 48.2082, 'lon': 16.3738},
    'Barcelona': {'lat': 41.3874, 'lon': 2.1686},
    'Berlin':    {'lat': 52.5200, 'lon': 13.4050},
    'Amsterdam': {'lat': 52.3676, 'lon': 4.9041},
}

# 2) Koordinaten hinzufügen
city_counts['lat'] = city_counts['city'].map(lambda c: city_coords[c]['lat'])
city_counts['lon'] = city_counts['city'].map(lambda c: city_coords[c]['lon'])

# 3) Überprüfung: Sind alle Angaben vollständig
assert city_counts['lat'].notna().all() and city_counts['lon'].notna().all(), "Städte ohne Koordinaten"

# Erstellen eines Plotly-Objekts auf Basis von Mapbox (Karte + Punkte).
fig = px.scatter_mapbox(
    city_counts,
    lat='lat',
    lon='lon',
    size='count',                          
    color='count',                        
    hover_name='city',                     
    hover_data={'count': True, 'lat': False, 'lon': False},
    size_max=50,
    zoom=3,
    mapbox_style='open-street-map',
    color_continuous_scale=mein_palette_hex,     # ← ваша палитра
    range_color=(city_counts['count'].min(), city_counts['count'].max())
)

#Layout-Einstellungen
fig.update_layout(
    title='Airbnb: Anzahl der Anzeigen nach Städten (Bubble Map)',
    margin=dict(l=0, r=0, t=40, b=0),
    legend_title_text='Anzahl der Anzeigen',
    mapbox_center={'lat': 46.0, 'lon': 8.0},
    mapbox_zoom=3,
    height=420,   
    width=650    
)

fig.show()
```
